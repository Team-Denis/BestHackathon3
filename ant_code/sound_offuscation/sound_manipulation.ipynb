```python
import scipy as sp
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile
```

```python
music = wavfile.read('data/oppa_input.wav')
```

```python
music
```

```python
fs = music[0]
sound_data = music[1]
```

```python
plt.plot(sound_data)
```

```python
length = (sound_data.shape[0] // fs)  # in seconds

length
```

```python
sound_data = sound_data[:fs * length]
```

```python
# scale the data
sound_data = sound_data / np.max(np.abs(sound_data))
```

```python
signal_data = np.sin(2 * np.pi * 200 * np.linspace(0, length, fs * length))
signal_data.shape, sound_data.shape
```

```python
signal_data
```

```python
sound_data = sound_data * 0.8 + 0.2 * signal_data
```

```python
plt.plot(sound_data)
```

```python
sound_data = sound_data * 32_767
wavfile.write("output/oppa_output.wav", fs, sound_data.astype(np.int16))
```
